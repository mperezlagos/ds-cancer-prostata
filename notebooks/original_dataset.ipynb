```python
import pandas as pd

# Cargar el dataset

df = pd.read_excel("../data/raw/cancer_hbv.xlsx", engine='openpyxl')

# Mostrar las primeras filas y un resumen de los datos
df.describe(), df.info()
```

```python
# 1. Obtener información sobre las columnas
diccionario_columnas = {}

for columna in df.columns:
    tipo_variable = str(df[columna].dtype)  # Convertir a cadena de texto
    
    # Agregar información sobre si la variable es numérica o categórica
    if pd.api.types.is_numeric_dtype(df[columna]):
        tipo_variable += " (Numérica)"
    else:
        tipo_variable += " (Categórica)"
    
    # Resumen del contenido
    if pd.api.types.is_datetime64_any_dtype(df[columna]):
        resumen_contenido = df[columna].describe(datetime_is_numeric=True).reset_index().rename(columns={'index': 'Descripción'})
    else:
        resumen_contenido = df[columna].describe().reset_index().rename(columns={'index': 'Descripción'})
    
    # Agregar al diccionario
    diccionario_columnas[columna] = {
        "Tipo": tipo_variable,
        "Resumen": resumen_contenido
    }
```

```python
# Agregar la impresión del diccionario_columnas aquí
print(diccionario_columnas)
```

```python
# 2. Obtener información sobre las columnas
diccionario_columnas_numericas = {}
diccionario_columnas_categoricas = {}

for columna in df.columns:
    tipo_variable = str(df[columna].dtype)  # Convertir a cadena de texto
    
    # Resumen del contenido
    if pd.api.types.is_datetime64_any_dtype(df[columna]):
        resumen_contenido = df[columna].describe(datetime_is_numeric=True).reset_index().rename(columns={'index': 'Descripción'})
    else:
        resumen_contenido = df[columna].describe().reset_index().rename(columns={'index': 'Descripción'})
    
    # Agregar información sobre si la variable es numérica o categórica
    if pd.api.types.is_numeric_dtype(df[columna]):
        tipo_variable += " (Numérica)"
        # Agregar al diccionario de columnas numéricas
        diccionario_columnas_numericas[columna] = {
            "Tipo": tipo_variable,
            "Resumen": resumen_contenido
        }
    else:
        tipo_variable += " (Categórica)"
        # Agregar al diccionario de columnas categóricas
        diccionario_columnas_categoricas[columna] = {
            "Tipo": tipo_variable,
            "Resumen": resumen_contenido
        }

# 3. Exportar el diccionario a Excel
ruta_excel = "C:/Users/Administrador/Desktop/Magister_DataScience/21_Proyecto_de_Grado_I/ds-cancer-prostata/data/processed/Diccionario_original.xlsx"   # Ruta donde se guardará el archivo Excel
with pd.ExcelWriter(ruta_excel) as writer:
    # Hoja para columnas numéricas
    df_numericas = pd.DataFrame()
    for columna, info in diccionario_columnas_numericas.items():
        resumen_transpuesto = info['Resumen'].T.reset_index()
        resumen_transpuesto['Columna'] = columna
        resumen_transpuesto = resumen_transpuesto.rename(columns={0: 'Descripción', columna: 'Valor'})
        df_numericas = pd.concat([df_numericas, resumen_transpuesto], ignore_index=True)
    df_numericas.to_excel(writer, index=False, sheet_name="Columnas Numéricas")
    
    # Hoja para columnas categóricas
    df_categoricas = pd.DataFrame()
    for columna, info in diccionario_columnas_categoricas.items():
        resumen_transpuesto = info['Resumen'].T.reset_index()
        resumen_transpuesto['Columna'] = columna
        resumen_transpuesto = resumen_transpuesto.rename(columns={0: 'Descripción', columna: 'Valor'})
        df_categoricas = pd.concat([df_categoricas, resumen_transpuesto], ignore_index=True)
    df_categoricas.to_excel(writer, index=False, sheet_name="Columnas Categóricas")

print("El diccionario de columnas se ha exportado a Excel exitosamente.")
```
